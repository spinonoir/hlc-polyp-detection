```python
import pandas as pd
import matplotlib.pyplot as plt
_Blue = '#2CBDFE'
_Green = '#47DBCD'
_Pink = '#F3A0F2'
_Purple = '#9D2EC5'
_Violet = '#661D98'
_Amber = '#F5B14C'
```

```python
run1_p = pd.read_csv("run1_precision.csv")
run1_r = pd.read_csv("run1_recall.csv")
run1_f1 = pd.read_csv("run1_f1.csv")
```

```python
run2_p = pd.read_csv("run2_precision.csv")
run2_r = pd.read_csv("run2_recall.csv")
run2_f1 = pd.read_csv("run2_f1.csv")
```

```python
run2_p["Step"] = run2_p["Step"] + 259183
run2_r["Step"] = run2_r["Step"] + 259183
run2_f1["Step"] = run2_f1["Step"] + 259183
```

```python

fig, ax1 = plt.subplots(1, 1, sharex=True, figsize=(10, 5))

# plot lines for precision, recall, and f1 score vs global step
ax1.plot(run1_p['Step'], run1_p['r1_val_precision'], label='run1 Precision', color=_Blue)
ax1.plot(run1_r['Step'], run1_r['r1_val_recall'], label='run1 Recall', color=_Green)
ax1.plot(run1_f1['Step'], run1_f1['r1_val_f1'], label='run1 F1 Score', color=_Pink)
ax1.plot(run2_p['Step'], run2_p['run2_val_precision'], label='run2 Precision', color=_Purple)
ax1.plot(run2_r['Step'], run2_r['run2_recall'], label='run2 Recall', color=_Amber)
ax1.plot(run2_f1['Step'], run2_f1['run2_val_f1'], label='run2 F1 Score', color="blue")
ax1.legend(loc='upper left')
# set the chart title and axis labels
plt.title('FasterRCNN Metrics')
plt.xlabel('Gloabl Steps')


# show the plot
plt.show()
```
